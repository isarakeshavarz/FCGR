##### import library

```python
import fcgr
import seaborn as sns
import pandas as pd
import numpy as np
```

##### define kmer length

```python
kmer_length = 10
```

##### generate FCGR key

```python
chaos_game_kmer_array = fcgr.chaos_game_representation_key(kmer_length=kmer_length)
```

##### provide file path

```python
file_path_human = "/home/abhishek/Desktop/gwas/GCF_000001405.25_GRCh37.p13_genomic.fna"
```

##### generate FCGR matrix

```python
print("length of sequence of human is: ",  len(fcgr.read_fasta(file_path_human)))
fcgr_sequence_human = fcgr.chaos_frequency_matrix(fasta_string= fcgr.read_fasta(file_path_human), kmer_length = kmer_length,  pseudo_count = True)
```

##### Plot the heatmap

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, dpi = 600, figsize =(3.5, 3.5))

heatmap = sns.heatmap(-np.log(fcgr_sequence_human[0]/fcgr_sequence_human[0].ravel().sum()), ax=ax, cmap= "inferno", cbar_kws={'orientation': 'vertical', 'pad': 0.02})
ax.set_xlabel(None)
ax.set_ylabel(None)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_yticks([])
ax.set_xticks([])
# ax.set_title("FCGR heatmap of $\it{H.sapien}$ ")
ax.tick_params(axis='x', labelsize=4, rotation=90)
ax.tick_params(axis='y', labelsize=4)
ax.tick_params(axis='both', which='both', width=0.5)
ax.xaxis.label.set_fontsize(4)
ax.yaxis.label.set_fontsize(4)
ax.title.set_fontsize(14)

cbar = heatmap.collections[0].colorbar
cbar.ax.yaxis.set_tick_params(labelsize=12)
plt.tight_layout()

plt.savefig("human_final.png", dpi = 600)
```
